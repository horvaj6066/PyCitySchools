```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# lets work out the district totals and averages and add these to the school_data dataframe 
math_pass=0
read_pass=0
student_count=school_data['size'].sum() 
school_count=school_data['school_name'].count()
total_budget=school_data['budget'].sum() 
avg_reading_score = student_data['reading_score'].sum()/student_data['reading_score'].count()
avg_math_score = student_data['math_score'].sum()/student_data['math_score'].count()
avg_score= (avg_math_score+avg_reading_score)/2


for j in student_data['math_score']:
    if j>=70:
        math_pass +=1
math_pass= math_pass/student_data['math_score'].count()        
        
for j in student_data['reading_score']:
    if j>=70:
        read_pass +=1
read_pass= read_pass/student_data['reading_score'].count()
        
#created the district summary dataframe 

district_summary = pd.DataFrame(index=[0])
district_summary['Number Schools'] = (school_count)
district_summary['Number Students'] = student_count
district_summary['Budget'] = "$"+ str(total_budget) 
district_summary['Avg math score'] = avg_math_score
district_summary['Avg reading score'] = avg_reading_score
district_summary['Avg score'] = avg_score
district_summary['% passing math'] = math_pass*100
district_summary['% passing reading'] = read_pass*100 

district_summary.round(2)


```

```python
#init variables to iterate thru the schools and calculate all the average scores by grade and other
grade_9_reading=[]
grade_9_math=[]
grade_10_reading=[]
grade_10_math=[]
grade_11_reading=[]
grade_11_math=[]
grade_12_reading=[]
grade_12_math=[]
high_school=[] #list of schools
budget_per_student=[]  #list for budget/student
schools=[]
school_avg_math_score = []
school_avg_reading_score = []
school_math_pass_rate=[]
school_reading_pass_rate=[]
school_avg_passing_rate=[]
schools=school_data['school_name']

high_school_counter=0
for i in schools:   # go thru the schools and pull out the grades 
    high_school.append(student_data.loc[student_data['school_name'] == i])  #add to high school all the data for particular school
    
    #pull out scores by grade for each high school as we iterate thru
    grades_by_grade = school_data_complete.loc[school_data_complete['school_name'] == i] #[grade'] == 9))
    
    grade_9_scores = grades_by_grade.loc[grades_by_grade['grade'] == '9th']
    grade_9_scores = grade_9_scores.groupby('school_name')['reading_score'].mean()
    grade_9_reading.append(grade_9_scores.iloc[0]) 
    #print(grade_9_scores.iloc[0]) #debug print
    
    grade_9_scores = grades_by_grade.loc[grades_by_grade['grade'] == '9th']
    grade_9_scores = grade_9_scores.groupby('school_name')['math_score'].mean()
    
    grade_9_math.append(grade_9_scores.iloc[0]) 
    #print(grade_9_scores.iloc[0])
    
    grade_10_scores = grades_by_grade.loc[grades_by_grade['grade'] == '10th']
    grade_10_scores = grade_10_scores.groupby('school_name')['reading_score'].mean()
    grade_10_reading.append(grade_10_scores.iloc[0]) 
    #print(grade_10_scores.iloc[0])
    
    grade_10_scores = grades_by_grade.loc[grades_by_grade['grade'] == '10th']
    grade_10_scores = grade_10_scores.groupby('school_name')['math_score'].mean()
    grade_10_math.append(grade_10_scores.iloc[0])
    #print(grade_10_scores.iloc[0])   

    grade_11_scores = grades_by_grade.loc[grades_by_grade['grade'] == '11th']
    grade_11_scores = grade_11_scores.groupby('school_name')['reading_score'].mean()
    grade_11_reading.append(grade_11_scores.iloc[0]) 
    #print(grade_11_scores.iloc[0]
    
    grade_11_scores= grades_by_grade.loc[grades_by_grade['grade'] == '11th']
    grade_11_scores = grade_11_scores.groupby('school_name')['math_score'].mean()
    grade_11_math.append(grade_11_scores.iloc[0])
    #print(grade_11_scores.iloc[0]
          
    grade_12_scores = grades_by_grade.loc[grades_by_grade['grade'] == '12th']
    grade_12_scores = grade_12_scores.groupby('school_name')['reading_score'].mean()
    grade_12_reading.append(grade_12_scores.iloc[0]) 
    #print(grade_12_scores.iloc[0]
          
    grade_12_scores= grades_by_grade.loc[grades_by_grade['grade'] == '12th']
    grade_12_scores = grade_12_scores.groupby('school_name')['math_score'].mean()
    grade_12_math.append(grade_12_scores.iloc[0])
    #print(grade_12_scores.iloc[0]
          
    #take the math & reading scores from each school and avg them and append 
    math_score = high_school[high_school_counter]['math_score'] #pull out all the math scores for each high school 
    school_avg_math_score.append(math_score.sum()/len(math_score))  #calc the avg and append to the list 
    reading_score = high_school[high_school_counter]['reading_score']  #pull out all the reading  scores for each high school 
    school_avg_reading_score.append(reading_score.sum()/len(reading_score))  #calc the avg and append to the list 
    
    school_math_pass=0  #reset
    for j in math_score:  #this loop is to run through and count how many math scores were passing scores
        if j>=70:
            school_math_pass +=1
    school_math_pass_rate.append(school_math_pass/len(math_score)*100) #calculate the pass rate and append to the list 
    
    school_reading_pass=0
    for j in reading_score:  #this loop is to run through and count how many reading scores were passing scores
        if j>=70:
            school_reading_pass +=1
    school_reading_pass_rate.append(school_reading_pass/len(reading_score)*100) #calculate the pass rate 
    school_avg_passing_rate.append((school_math_pass_rate[high_school_counter] + school_reading_pass_rate[high_school_counter])/2)
       
    high_school_counter+=1 #increment the high school counter 
    budget_per_student = school_data['budget']/school_data['size']

#add school summary data
school_data['budget/student'] = budget_per_student 
school_data['avg math score'] = school_avg_math_score
school_data['avg reading score'] = school_avg_reading_score
school_data['math pass rate'] = school_math_pass_rate
school_data['reading pass rate'] = school_reading_pass_rate
school_data['avg pass rate'] = school_avg_passing_rate

#create a dataframe to store and display these variables
school_summary=school_data.drop(['School ID'], axis=1)
school_summary.round(2)
```
